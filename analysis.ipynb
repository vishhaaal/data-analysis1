```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Date = df['Time_stamp']
M01AB = df['M01AB']
M01AE = df['M01AE']
N02BA = df['N02BA']
N02BE = df['N02BE']
N05B = df['N05B']
N05C = df['N05C']
R03 = df['R03']
R06 = df['R06']

plt.rcParams['figure.figsize'] = (14,9)
plt.rcParams['date.converter'] = 'concise'
plt.rcParams['date.interval_multiples'] = True

plt.plot(Date,M01AB,color = 'tab:purple',label = 'M01AB')
plt.plot(Date,M01AE,color = 'tab:pink',label = 'M01AE')
plt.plot(Date,N02BA,color = 'tab:blue',label = 'N02BA')
plt.plot(Date,N02BE,color = 'tab:green',label = 'N02BE')
plt.plot(Date,N05B,color = 'tab:red',label = 'N05B')
plt.plot(Date,N05C,color = 'tab:orange',label = 'N05C')
plt.plot(Date,R03,color = 'tab:cyan',label = 'R03')
plt.plot(Date,R06,color = 'tab:olive',label = 'R06')

plt.legend()
plt.title("BIG PHARMA SALES")
plt.xlabel('DATE')
plt.ylabel('SALES NUMBERS')

plt.tight_layout()
plt.show
```

```python
df['Total'] = df['M01AB'] + df['M01AE'] + df['N02BA'] + df['N02BE'] + df['N05B'] + df['N05C'] + df['R03'] + df['R06']
sns.pairplot(data=df)
```
